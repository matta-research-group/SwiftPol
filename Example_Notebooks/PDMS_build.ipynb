```python
# Build short PDMS chain (adjust num_chains to get a larger ensemble)
# Ignore OpenEye warnings, license not needed for this application
# Si(C)(C)(C) caps
sys = build.polymer_system_from_PDI(
            monomer_list=['I-[Si](-C)(-C)-O-I'],
            reaction='[O:1][I:3].[Si:2][I:4]>>[O:1][Si:2].[I:3][I:4]',
            length_target=8,
            terminals='C',
            initiator='[Si](-C)(-C)(-C)',  
            num_chains=1,
            PDI_target=1.0,
            acceptance=5
        )
```

```python
sys.chain_rdkit[0]
```

```python
sys.generate_conformers()
```

```python
chain_rdmol = Chem.AddHs(sys.chain_rdkit[0])
Chem.MolToPDBFile(chain_rdmol, 'chain.pdb')
chain_rdmol
```
